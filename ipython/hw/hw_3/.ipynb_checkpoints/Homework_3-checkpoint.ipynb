### Part 1: Data preparation
Create a data preparation and cleaning function that does the following:
- Has a single input that is a file name string
- Reads data (the data is comma separated, has a row header and the first column `EventID` is the index) into a pandas `dataframe`
- Cleans the data
  - Convert the feature `Label` to numeric (choose the minority class to be equal to 1)
    - Create a feature `Y` with numeric label
    - Drop the feature `Label`
  - If a feature has missing values (i.e., `-999`): 
    - Create a dummy variable for the missing value
      - Call the variable `orig_var_name` + `_mv` where `orig_var_name` is the name of the actual var with a missing value
      - Give this new variable a 1 if the original variable is missing
    - Replace the missing value with the average of the feature (make sure to compute the mean on records where the value isn't missing). You may find pandas' `.replace()` function useful.
- After the above is done, rescales the data so that each feature has zero mean and unit variance (hint: look up sklearn.preprocessing)
- Returns the cleaned and rescaled dataset

Hint: as a guide, this function can easily be done in less than 15 lines.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
def cleanBosonData(file_path): 
    #reads the file from the source and creates a pandas dataframe
    df = pd.read_csv(file_path,index_col = 0)
    #replace the attribute "Label" with 0 and 1 as mentioned in the question
    df['Y'] = df['Label'].replace(['s','b'],[1,0])
    df = df.drop('Label',1)
    orig_cols = df.drop('Y',1)
    for x in orig_cols.columns:
        #replacing all the -999 with NAn
        df[x] = df[x].replace([-999],[np.NaN])
        if(df[x].isnull().values.any()):
            df[x + '_mv'] = np.isnan(df[x])*1
        df[x] = df[x].replace([np.NaN], df[x].mean(axis=0))
        scale = StandardScaler()
        df[x] = scale.fit_transform(df[x],df['Y']) 
    return df
```

For each model above, plot the ROC curve of both models on the same plot. Make sure to use the test set for computing and plotting. In the legend, also print out the Area Under the ROC (AUC) for reference.

```python
import matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC
from sklearn import metrics
%matplotlib inline
#Logistic Regression Classifier
LR = LogisticRegression(C=10^30)
#SVM Classifier
SVM = SVC(kernel='linear', probability=True)
#Training and building the model
LR = LR.fit(data_train.drop('Y',1),data_train['Y'])
SVM = SVM.fit(data_train.drop('Y',1),data_train['Y'])
#Predicting the score
a=data_test.drop('Y', 1)
predict_LR = LR.predict_proba(a)
predict_SVM = SVM.predict_proba(a)
false_positive_LR, true_positive_LR, thresholds = metrics.roc_curve(data_test['Y'], predict_LR[:, 1])
false_positive_SVM, true_positive_SVM, thresholds = metrics.roc_curve(data_test['Y'], predict_SVM[:, 1])
roc_auc_LR = metrics.auc(false_positive_LR, true_positive_LR)
roc_auc_SVM = metrics.auc(false_positive_SVM, true_positive_SVM)
plt.figure(1)
plt.plot(false_positive_LR, true_positive_LR, label='LR, AUC: {}'.format(roc_auc_LR)) 
plt.plot(false_positive_SVM, true_positive_SVM, label='SVM, AUC: {}'.format(roc_auc_SVM)) 
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('FPR-->')
plt.ylabel('TPR-->')
plt.title('Receiver Operator Characteristics curve')
plt.legend(loc='best')
plt.show()
```

### Part 3: Model selection with cross-validation
We think we might be able to improve the performance of the SVM if we perform a grid search on the hyper-parameter $C$.  Because we only have 1000 instances, we will have to use cross-validation to find the optimal $C$.

1\. Write a cross-validation function that does the following:
- Takes as inputs a dataset, a label name, # of splits/folds (`k`), a sequence of values for $C$ (`cs`)
- Performs two loops
  - Outer Loop: `for each f in range(k)`:
    - Splits the data into `data_train` & `data_validate` according to cross-validation logic
  - Inner Loop: `for each c in cs`:
    - Trains an SVM on training split with `C=c, kernel="linear"`
    - Computes AUC_c_k on validation data
    - Stores AUC_c_k in a  dictionary of values
- Returns a dictionary, where each key-value pair is: `c:[auc-c1,auc-c2,..auc-ck]`

```python
#importing all the necessary python libraries
from sklearn.metrics import roc_curve, auc
from sklearn.svm import SVC
from collections import defaultdict
import collections
def xValSVM(dataframe, label_name, k, cs):
    aucs = defaultdict(list)
    for f in range(k):
        data_val = dataframe.iloc[int(np.floor(f*len(dataframe)/k)):int(np.ceil((f+1)*len(dataframe)/k))]
        data_train = dataframe[~dataframe.isin(data_val)].dropna()
        for i, c in enumerate(cs):
            clf_SVM = SVC(kernel='linear',C=c,probability=True)
            pred = clf_SVM.fit(data_train.drop(label_name,1),data_train[label_name]).predict_proba(data_val.drop(label_name,1))
            fpr, tpr, thresholds = roc_curve(data_val[label_name], pred[:, 1])
            AUC_c_k=auc(fpr, tpr)
            aucs[c].append(AUC_c_k)
    aucs = collections.OrderedDict(sorted(aucs.items()))
    return aucs
```
